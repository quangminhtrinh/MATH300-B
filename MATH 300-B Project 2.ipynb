(3) The fifth Legendre polynomial is $P_5(x) = x^5 - \frac{10}{9}x^3 + \frac{5}{21}x.$

```julia
f(x) = x^5 - (10.0/9.0) * x^3 + (5/21) * x;
df(x) = 5.0 * x^4 - (10.0/3.0) * x^2 + (5.0/21.0);

# 1. Apply Newton method

function newton(f, df, p0, n_max=100, rel_tol=1e-8; verbose = true)
    converged = false;
    p = p0;
    p_old = p0;
    
    for i in 1:n_max

        p = p_old - f(p_old) / df(p_old)

        if verbose
            println("$i: p = $(p), |f(p)| = $(abs(f(p)))")
        end
        
        if (i>1)
            if abs(p-p_old)/abs(p)< rel_tol
                converged = true;
                break
            end
        end

        p_old = p;
        
    end
    
    if !converged
        print("ERROR: Did not converge after $n_max iterations")
    end

    return p
    
end

# 2. Find all five roots (nodes)

# Root guessing rationale:
# f(x) roots are symmetric around 0 and lie in (-1, 1). 
# Due to clustering (Chebyshev-like distribution), I guess: 
# - One root at x=0 (P5(0)=0).
# - Two positive roots near 0.5 (inner) and 0.9 (outer).

# Initial guesses based on the approximate closed-form solution:
# Smallest positive root
x_p1 = newton(f, df, 0.5)

# Largest positive root
x_p2 = newton(f, df, 0.9)

# The five nodes (including the symmetric negative roots and zero)
nodes = sort([-x_p2, -x_p1, 0.0, x_p1, x_p2])

# 3. Output results
print("\nGaussian Quadrature Nodes (n=5)")
println("\nRoots of P5(x) with relative error < 1e-8:")
for i in 1:length(nodes)
    node = nodes[i]
    println("x[$i] = $node") 
end

print("\nf(x) value at roots:\n")
for node in nodes
    println("f($node) = $(f(node))") 
end
```

(b) Using the computed nodes, apply Theorem 4.7 to compute the weights. Compare with Table 4.12 from the textbook. Do NOT use generative AI to complete this part of the assignment.

```julia
using QuadGK

# 1. Compute the Lagrange polynomial L_i(x)
function lagrange_polynomial(x, x_nodes, i)
    x_i = x_nodes[i]
    product = 1.0
    for j in 1:length(x_nodes)
        # Skip the j=i term
        if i != j
            x_j = x_nodes[j]
            product *= (x - x_j) / (x_i - x_j)
        end
    end
    return product
end

# 2. Compute the Gaussian Quadrature weight c_i by integrating the L_i(x) over [-1, 1]
function compute_weight(x_nodes, i)
    # Define the Lagrange basis polynomial L_i(x) as a function of x
    L_i(x) = lagrange_polynomial(x, x_nodes, i)
    
    # Integrate L_i(x) using QuadGK     
    integral, err = quadgk(L_i, -1.0, 1.0, rtol=1e-8)
    return integral
end

# 3. Store weights using an external loop
weights = zeros(length(nodes))

# Loop to compute and store all weights
for i in 1:length(nodes)
    # Call the existing function for a single weight
    c_i = compute_weight(nodes, i)
    weights[i] = c_i 
end
```

(4) Combining Algorithms: The Quantile FunctionThe goal of this problem is to numerically evaluate the inverse of a Cumulative Distribution Function (CDF) (the quantile function) for the standard normal distribution.Recall that the CDF, in probabilistic terms, is:$$F(x) = \text{CDF}(x) = P(X \le x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi}}e^{-t^{2}/2}dt$$To simplify, recall that since the standard normal distribution is symmetric, $F(0) = 1/2$. Hence, the function is:$$F(x) = \text{CDF}(x) = \frac{1}{2} + \int_{0}^{x} \frac{1}{\sqrt{2\pi}}e^{-t^{2}/2}dt$$And, as noted, $F^{-1}(p)$ is the quantile function. That is to say, the 90% percentile (for example) is the value of $x$ for which $F(x) = 0.9$. For simplicity, we will assume $x \ge 0$ in this problem.

a) Using Composite Simpson's rule, write a function that returns $F(x)$ and $F'(x)$ at a given $x \ge 0$. The absolute error of composite Simpson's should be smaller than $10^{-8}$.

HINT: This should be of the form:


```julia
function F_comp_simpsons(x, n)
    # Fill in code here
    return F, Fp
end
```

where the outputs, $F$ and $Fp$, are:$$F(x) = \frac{1}{2} + \int_{0}^{x} \frac{1}{\sqrt{2\pi}}e^{-t^{2}/2}dt \quad \text{and} \quad F'(x) = \frac{1}{\sqrt{2\pi}}e^{-x^{2}/2}$$Do NOT use generative AI to complete this part of the assignment.

```julia
# 1. PDF Function
f(t) = (1.0 / sqrt(2 * pi)) * exp(-t^2 / 2)

# Let K = 1/sqrt(2 * pi)
# So, f(t) = K * exp(-t^2/2)

# Fourth derivative of f(t):
# f'(t) = -t * f(t) * K
# f''(t) = (t^2 - 1) * f(t) * K
# f'''(t) = (3t - t^3) * f(t) * K
# f^(4)(t) = (t^4 - 6t^2 + 3) * f(t) * K

# Let M4 be the max value of |f^(4)(t)| on [0, x]
# Apply the EVT to critical points 
# - t = 0
# - t = sqrt(5 + sqrt(10))
# - t = sqrt(5 - sqrt(10)) 
# M4 = max |f^(4)(t = 0)| = 3/sqrt(2 * pi)

# 2. Function F_comp_simpsons(x, n)
# Calculate F and F_prime for the CDF using Composite Simpson's rule
# n subintervals needs to satisfy error bound |E| < 1e-8

# Error bound |E| <= (b - a)/180 * h^4 * M4 
# h = (b - a) / n
# n >= ((x^5 * M4) / (180 * abs_tol))^(1/4)

function F_comp_simpsons(x)

    # 3. Handle the trivial case F(0)
    # Although the problem assumes x >= 0, consider x <= 0 to also handle x = 0
    if x <= 0
        return 0.5, f(0) # F(0) = 1/2
    end

    # 4. Determine n for |E| < 1e-8
    abs_tol = 1e-8
    M4 = 3 / sqrt(2 * pi)
    n = ((x^5 * M4) / (180 * abs_tol))^(1/4)

    # n must be a positive, even integer
    n = ceil(Int, n)
    if isodd(n)
        n += 1
    end
    
    if n < 2
        n = 2
    end
    
    # 5. Composite Simpson's rule integration
    a = 0
    b = x
    h = (b - a) / n
    
    # Integral sum = f(a) + f(b)
    integral_sum = f(a) + f(b) 
    
    # Add all interior points with coefficients 4, 2, 4, 2, ...
    for i in 1:(n - 1)
        x_i = a + i * h;
        
        # Coefficients in Composite Simpson's integration
        # - 4 for odd index (1, 3, 5, ...)
        # - 2 for even index (2, 4, 6, ...)
        if isodd(i)
            integral_sum += 4 * f(x_i);
        else
            integral_sum += 2 * f(x_i);
        end
    end
    
    # Integral = h/3 * Integral sum
    integral = (h / 3) * integral_sum;
    
    # F(x) = 1/2 + Integral
    F = 0.5 + integral
    
    # Final F'(x) = f(x)
    F_prime = f(x)
    
    return F, F_prime
end
```

(b) Using Newton’s method with $F_{\text{comp\_simpsons}}$ from (a), write the function $\text{invF}(p)$ which approximates $F^{-1}(p)$. The estimated relative error of Newton’s method should be smaller than $10^{-8}$. Do NOT use generative AI to complete this part of the assignment.

```julia
# 1. Newton's Method for Inverse CDF (Refactored to define G(x) explicitly)
function invF(p_target, n_max=100, rel_tol=1e-8; verbose = false)
    
    # F(x) = CDF value (first element of tuple)
    F(x) = F_comp_simpsons(x)[1]
    
    # Fp(x) = F'(x) = PDF value (second element of tuple)
    Fp(x) = F_comp_simpsons(x)[2]

    # G(x) = F(x) - p_target
    G(x) = F(x) - p_target
    
    # Gp(x) = G'(x) = F'(x)
    Gp(x) = Fp(x)
    
    # Since we're only concerned with p >= 0.5 (or x >= 0)
    if p_target < 0.5 || p_target >= 1.0
        println("ERROR: Input probability p out of range. Must be in [0.5, 1.0).")
    end

    # 2. Initial guess near 0 p0 = 0.1 
    x_old = 0.1; 
    x_current = x_old;
    converged = false;

    for i in 1:n_max
        
        # 3. Calculate G(x_old) and G'(x_old)
        G_current = G(x_old); 
        Fp_current = Gp(x_old); # This is G'(x_old)

        # 5. Safeguard against division by zero (G'(p_current) = F'(p_current))
        if abs(Fp_current) < 1e-8 
             println("Warning: Derivative F'(x) is near zero. Stopping Newton's method.");
             break;
        end

        # 6. Newton's formula: x_current = x_old - G(x_old) / G'(x_old)
        x_current = x_old - G_current / Fp_current;
        
        if verbose
             # For verbose output, retrieve F_current from G_current and p_target
             F_current = G_current + p_target
             println(" $i: x = $(x_current), |F(x) - p| = $(abs(G_current))");
        end
        
        if (i>1)
            # 5. Check estimated relative error
            if abs(x_current) > 1e-8 && abs(x_current - x_old) / abs(x_current) < rel_tol
                converged = true;
                break;
            end
        end

        # 7. Update the approximation for the next iteration
        x_old = x_current;
    end
    
    if !converged
        println("ERROR: Newton's method failed to converge after $n_max iterations for p_target=$p_target.");
    end

    return x_current;
end
```

```julia
using Roots, QuadGK

# 1. PDF Function
f(t) = (1.0 / sqrt(2 * pi)) * exp(-t^2 / 2)

# 2. CDF Function F(x) using QuadGK
# Compute the CDF by integrating the PDF from -Infinity to x 
function F_quadgk(x)    
    integral, err = quadgk(f, -Inf, x, rtol=1e-8)
    return integral
end

# 3. Inverse CDF Function using Roots.jl
function invF_quadgk(p_target)
    # Since we're only concerned with p >= 0.5 (or x >= 0)
    if p_target < 0.5 || p_target >= 1.0
        println(ERROR:"Input probability p out of range.")
    end

    # 4. G(x) = F(x) - p_target
    # F(x) implemented via F_quadgk(x)
    G(x) = F_quadgk(x) - p_target

    # 5. Use Roots.jl to find the root x where G(x) = 0
    # We use a bracket [0.0, 3.0] for initial search.
    x_quantile = find_zero(G, [0.0, 3.0])
    return x_quantile
end
```
